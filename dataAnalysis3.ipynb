```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(
    df.iloc[:, 3:].values,
    df[['dpi', 'Sana', 'Tratamiento']],
    test_size=0.2,
    random_state=87
)
```

```python
print(f"train size: {y_train.shape[0]}\n test size: {y_test.shape[0]}")
```

```python
from sklearn.svm import SVC
best_params = {'C': 100000, 'class_weight': None, 'decision_function_shape': 'ovr', 'kernel': 'linear'}
clf = SVC(**best_params)
```

```python
from dataFunction3 import kfoldCV_clf, model_confmat, learningCurve_clf
```

```python
kfoldCV_clf(clf, X_train, y_train['Sana'].values, n_splits=7)
```

```python
clf.fit(X_train, y_train['Sana'].values)
model_confmat(clf, X_test, y_test['Sana'].values)
```

```python
columnas = df.columns.to_list()[3:]
columnas = list(map(int, columnas))
# coef es el valor absoluto de los coeficientes de mi estimador SVC
coef = pd.DataFrame(np.abs(clf.coef_).T, columns=['Pesos -1', 'Pesos 0', 'Pesos 1'], index=columnas)
coef.index.name = 'Longitud de Onda'
coef['Pesos Promedio'] = (1/3)*(coef['Pesos -1'] + coef['Pesos 0'] + coef['Pesos 1'])
coef
```

```python
# Crear la figura
plt.figure(figsize=(10, 6))
plt.plot(coef.index, coef['Pesos Promedio'], marker='o', linestyle='None', color='b')

# Títulos y etiquetas
plt.style.use('seaborn-v0_8-whitegrid')
plt.title('Importancia Longitud Onda')
plt.xlabel('Longitud de onda')
plt.ylabel('Peso')

# Rango del eje x
plt.xlim(200, 2650)

# Cuadrícula opcional
plt.grid(True)

# Mostrar la gráfica
plt.show()
```

## Análisis de las predicciones equivocadas

```python
y_pred = clf.predict(X_test)
y_test['Sana predict'] = y_pred
y_test['Match predict'] = (y_pred == y_test['Sana'])
y_test
```

```python
y_test = y_test[(y_test['Match predict'] == False)]
```

```python
y_test
```

```python
dpi_counts = y_test.groupby('dpi')['Match predict'].count()

plt.figure(figsize=(10, 5))
plt.bar(dpi_counts.index, dpi_counts.values, color='skyblue', alpha=0.7)

plt.xlabel('dpi')
plt.ylabel('Frecuencia')
plt.title('Frecuencia de dpi con Match predict = False')
plt.xticks(dpi_counts.index)  # Asegurar que los ticks coincidan con los dpi presentes
plt.grid(axis='y', linestyle='--', alpha=0.7)

plt.show()
```

```python
Tratamiento_counts = y_test.groupby('Sana')['Match predict'].count()

plt.figure(figsize=(10, 5))
plt.bar(Tratamiento_counts.index, Tratamiento_counts.values, color='skyblue', alpha=0.7)

plt.xlabel('Tratamiento')
plt.ylabel('Frecuencia')
plt.title('Frecuencia de Sana con Match predict = False')
plt.xticks(Tratamiento_counts.index)  # Asegurar que los ticks coincidan con los dpi presentes
plt.xticks(rotation=0)
plt.grid(axis='y', linestyle='--', alpha=0.7)

plt.show()
```

```python
Tratamiento_counts = y_test.groupby('Tratamiento')['Match predict'].count()

plt.figure(figsize=(10, 5))
plt.bar(Tratamiento_counts.index, Tratamiento_counts.values, color='skyblue', alpha=0.7)

plt.xlabel('Tratamiento')
plt.ylabel('Frecuencia')
plt.title('Frecuencia de Tratamiento con Match predict = False')
plt.xticks(Tratamiento_counts.index)  # Asegurar que los ticks coincidan con los dpi presentes
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.7)

plt.show()
```
